```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set Chrome options
options = Options()
options.add_argument("--disable-gpu")  # Disable GPU acceleration

# Set ChromeDriver logging options
service_args = ['--verbose']  # Set any desired logging options here

# Create ChromeDriver service
service = Service(ChromeDriverManager().install(), service_args=service_args)

# Pass service to webdriver
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()

url = "https://www.realtor.ca/realtors/canada" 

driver.get(url)
time.sleep(60)  # Allow time for the page to load

results = driver.find_element(By.CSS_SELECTOR, "span.realtorResultsTotalResultsNum").get_attribute('innerHTML')

if results.strip():
    results = int(results.replace(',', ''))
else:
    print("Error: No results found")

numOfPages = (( results + 19 ) // 20) + 1

# Loop over the pages
for page_num in range(1, numOfPages):
    url = f'https://www.realtor.ca/realtor-search-results#sort=11-A&v=location&page={page_num}'

    driver.get(url)
    time.sleep(20)  # Allow time for the page to load
            
    # Wait for the specific elements to load
    elements = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div#realtorSearchResultsCardsCon > div.realtorSearchResultCardCon"))
    )

    for element in elements:
        name_tag = element.find_element(By.CSS_SELECTOR, 'span.realtorCardName').get_attribute('innerHTML')
        name = name_tag if name_tag else 'N/A'

        # Extract company
        company_tag = element.find_element(By.CSS_SELECTOR, 'div.realtorCardOfficeName').get_attribute('innerHTML')
        company = company_tag if company_tag else 'N/A'

        # Extract phone
        phone_tag = element.find_element(By.CSS_SELECTOR, 'span.realtorCardContactNumber.TelephoneNumber').get_attribute('innerHTML')
        phone = phone_tag if phone_tag else 'N/A'

        # Extract address
        address_tag = element.find_element(By.CSS_SELECTOR, 'div.realtorCardOfficeAddress').get_attribute('innerHTML')
        address = address_tag if address_tag else 'N/A'

        # Extract website
        website_tag = element.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
        website = website_tag if website_tag else 'N/A'

        print(f"Name: {name}")
        print(f"Company: {company}")
        print(f"Phone: {phone}")
        print(f"Address: {address}")
        print(f"Website: {website}")
                    
        # Wait for 1/4 second before the next iteration
        time.sleep(0.25)
    
driver.quit()
```
